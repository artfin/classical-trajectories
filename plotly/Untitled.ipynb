```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.plotly as py
import json

# loading jx,jy,jz values
with open('simple-water.dat', mode = 'r') as inputfile:
    data = json.load(inputfile)
```

```python
theta = np.linspace(0, 2*pi, 100)
phi = np.linspace(0, pi, 100)
J = 25.
x = J * np.outer(np.cos(theta), np.sin(phi))
y = J * np.outer(np.sin(theta), np.sin(phi))
z = J * np.outer(np.ones(100), np.cos(phi))
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection = '3d')
ax.plot_surface(x,y,z, alpha = 0.1)
ax.plot(data['jx'], data['jy'], data['jz'], color = 'b')
plt.draw()
plt.show()
```

```python
import plotly.graph_objs as go
data = Data([
    Surface(
        x=x,
        y=y,
        z=z
    )
])

layout = Layout(
    title='Bloch sphere',
    autosize=False,
    width=500,
    height=500,
    images= [dict(opacity= 0.1)]
)

fig = Figure(data=data, layout=layout)
iplot(fig, filename='bloch-sphere-surface')
```
